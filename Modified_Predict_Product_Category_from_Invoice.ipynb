```
import zipfile
zip_ref = zipfile.ZipFile('4eddd640-9-dataset.zip', 'r')
zip_ref.extractall('MODEL/')
zip_ref.close()
```

```
import pandas as pd
train_csv=pd.read_csv("MODEL/dataset/Train.csv")
```

```
train_csv.head()
```

```
x_des=train_csv['Item_Description']
x_des.head()
```

```
x_des[0].split('-')[-1]
```

```

l=[]
for i in range(0,5719):
  l.append(x_des[i].split('-')[-1])
l
  
```

```
x_des_df = pd.DataFrame({'Item_Description':l})
x_des_df.head()
```

```
x_train=train_csv.drop(['Item_Description','Product_Category'],axis=1)
x_train['Item_Description']=x_des_df['Item_Description']
y_train=train_csv['Product_Category']
y_train=y_train.to_frame(name='Product_Category')
```

```
x_train.head()
```

```


categorical_feature_mask = x_train.dtypes==object
categorical_cols = x_train.columns[categorical_feature_mask].tolist()

```

```
from sklearn.preprocessing import LabelEncoder
# instantiate labelencoder object
le = LabelEncoder()
```

```
x_train[categorical_cols] = x_train[categorical_cols].apply(lambda col: le.fit_transform(col))
x_train[categorical_cols].head(10)
```

```
x_train
```

```
encoder = LabelEncoder()
encoder.fit(y_train['Product_Category'])
y_train['Product_Category_encoded'] = encoder.transform(y_train['Product_Category'])
```

```
Y_train=y_train.drop(['Product_Category'],axis=1)
```

```
Y_train
```

```
x_train.head()
```

```
test_csv=pd.read_csv("MODEL/dataset/Test.csv")
```

```
test_csv.head()
```

```
x_test=test_csv.drop('Item_Description',axis=1)
```

```
x_des_test=test_csv['Item_Description']
x_des_test.head()
x_des_test.shape
```

```
lt=[]
for i in range(0,2292):
  lt.append(x_des_test[i].split('-')[-1])
lt
```

```
x_des_test_df = pd.DataFrame({'Item_Description':lt})
x_des_test_df.head()
```

```
x_test['Item_Description']=x_des_test_df['Item_Description']
```

```
categorical_feature_mask_test = x_test.dtypes==object
categorical_cols_test = x_test.columns[categorical_feature_mask_test].tolist()
```

```
x_test[categorical_cols_test] = x_test[categorical_cols_test].apply(lambda col: le.fit_transform(col))
x_test[categorical_cols_test].head(10)
```

```
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
rf.fit(x_train, Y_train)
predicted_classes = rf.predict(x_test)
print(predicted_classes)
```

```
print(y_pred)
y_pred_df = pd.DataFrame({'Product_Category':y_pred})
```
